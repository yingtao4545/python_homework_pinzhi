```python

对上证指数进行如下操作：
1. 计算每日涨跌幅（用收盘价进行计算），并对其进行简单的描述行统计
2. 找出最大涨幅和跌幅对应的日期
3. 新建一列，统计指数表现:
    #涨跌幅 > 75% 分位数（六位小数），则为： 优秀
    #25% 分位数< zhangdiefu <=75% 分位数，则为：一般
    #涨跌幅《=25% 分位数，较差
```

```python
import pandas as pd
```

```python
index_df = pd.read_csv('/Users/wenjingli/Desktop/python_data_analysis/02_Pandas/02-1_PandasBasic/sh000001.csv', parse_dates=['candle_end_time'])
index_df.head()
                                      
```

```python
#change to label index
index_df.set_index('candle_end_time',inplace = True)
index_df
```

```python
#sort the label index
sorted_index=index_df.sort_index()
sorted_index
```

```python
# 1.1 add: columns= percentage_change = close/open -1; this way saves calculation
sorted_index['percentage_change'] = sorted_index['close']/sorted_index['open'] - 1
sorted_index
```

```python
#1.2 描述性统计
sorted_index['percentage_change'].describe()
```

```python
#2.1最大涨幅
maxincrease_date = sorted_index['percentage_change'].idxmax() #return: Timestamp('1994-08-10 00:00:00')
max_increase = sorted_index.loc[maxincrease_date,'percentage_change']
print(f'The max increase date is {maxincrease_date}, the increase is {max_increase}')
```

```python
#2.2 最大跌幅
maxdecrease_date = sorted_index['percentage_change'].idxmin() #return: Timestamp('1994-08-09 00:00:00')
max_decrease = sorted_index.loc[maxdecrease_date,'percentage_change']
print(f'The max decrease date is {maxdecrease_date}, the decrease is {max_decrease}')
```

```python
#3 categories，使用之前的describe 结果
# min        -0.131313
# 25%        -0.006465
# 50%         0.000578
# 75%         0.007436
# max         0.194271
group_names = ['优秀', '一般', '较差']
bins= [-0.131313, -0.006465, 0.007436,0.194271 ]
cats = pd.cut(sorted_index['percentage_change'], bins, labels= group_names)
sorted_index['groups'] = cats
sorted_index
```
